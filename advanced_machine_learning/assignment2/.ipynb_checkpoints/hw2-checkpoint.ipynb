```python
import pickle

tournaments = pickle.load(open('tournaments.pkl', 'rb'))
results = pickle.load(open('results.pkl', 'rb'))
```

```python
def get_train_tournaments_id():
    tournaments_2019 = [v for k,v in tournaments.items() if v['dateStart'][:4] == '2019']
    tournaments_2019_with_results = [v for idx, v in enumerate(tournaments_2019) if v['id'] in results and results[v['id']] != []]
    tournaments_2019_with_results_mask = [v for idx, v in enumerate(tournaments_2019_with_results) if 'mask' in results[v['id']][0]]
    return [v['id'] for v in tournaments_2019_with_results_mask]

def get_test_tournaments_id():
    tournaments_2020 = [v for k,v in tournaments.items() if v['dateStart'][:4] == '2020']
    tournaments_2020_with_results = [v for idx, v in enumerate(tournaments_2020) if v['id'] in results and results[v['id']] != []]
    tournaments_2020_with_results_mask = [v for idx, v in enumerate(tournaments_2020_with_results) if 'mask' in results[v['id']][0]]
    return [v['id'] for v in tournaments_2020_with_results_mask]
```

```python
train_tournaments_id = get_train_tournaments_id()
test_tournaments_id = get_test_tournaments_id()

print(len(train_tournaments_id))
print(len(test_tournaments_id))

tournaments_id = train_tournaments_id + test_tournaments_id
```

```python
def get_trimmed_results(results, ids):
  trimmed_results = {}
  for v in ids:
    trimmed_results[v] = results[v]
  return trimmed_results

results = get_trimmed_results(results, tournaments_id)
```

```python
import pandas as pd
import numpy as np


def get_dataset(tournaments_id):
    dataset = []
    for tournament_id in tournaments_id:
        teams_members = []
        masks = []        
        for team in results[tournament_id]:
            masks.append(team['mask'])
            team_members = []
            for player in team['teamMembers']:
                team_members.append(player['player']['id'])
            teams_members.append(team_members)
            
        complexities = np.zeros(len(masks[0]))
        for idx, mask in enumerate(masks):
            for question_num, result in enumerate(mask):
                complexities[question_num] += int(str(result))
        
        #print(complexities)   
        team_num = len(results[tournament_id])
        complexities = complexities / team_num
        #print(complexities) 
        
        #print(team_num)
        for complexity in complexities:
            complexity = complexity / team_num
        for question_idx, complexity in enumerate(complexities):
            for team_idx,team in enumerate(teams_members):
                result = masks[team_idx][question_idx]
                for player in team:                    
                    dataset.append([player, complexity, int(str(result))])
        return dataset
    

def vectorize(dataset, player_id_columns):
    X = []
    for row in dataset:
        one_hot_player = [0] * len(player_id_columns) #np.zeros(len(player_id_columns))
        player_idx = player_id_columns.index(row[0])
        one_hot_player[player_idx] = 1.0
        one_hot_player.append(row[1])
        X.append(one_hot_player)
    return X


def get_all_players_id(tournaments_id):
    ids = []
    for tournament_id in tournaments_id:
        for team in results[tournament_id]:
            for player in team['teamMembers']:
                ids.append(player['player']['id'])
    return sorted(list(set(ids)))




player_ids = get_all_players_id(train_tournaments_id) 

train_dataset = get_dataset(train_tournaments_id)
print(len(train_dataset))
#train_dataset = train_dataset[:5000]
X_train = vectorize(train_dataset, player_ids)
y_train = [train_dataset[i][2] for i in range(len(train_dataset))]  
```

```python
from sklearn.linear_model import LinearRegression#LogisticRegression

logisticRegr = LinearRegression() #(penalty='l1', solver='liblinear')
logisticRegr.fit(X_train, y_train)
#predictions = logisticRegr.predict(x_test)
print(logisticRegr.coef_, logisticRegr.intercept_)
```

```python
print(logisticRegr.coef_)
for idx, val in enumerate(logisticRegr.coef_[0]):
  if val > 0:
    print(idx, val)
```
